```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"

df = pd.read_csv(URL)
df.head()
```

```python
sns.catplot(x='FlightNumber', y='PayloadMass', hue='Class', data=df, aspect =5)
plt.xlabel("FlightNumber")
plt.ylabel("PayLoadMass")
plt.show
```

```python
sns.catplot(x='FlightNumber', y='LaunchSite', hue='Class', data=df)
plt.xlabel("FlightNumber")
plt.ylabel("LaunchSite")
plt.show
```

```python
sns.scatterplot(x='FlightNumber',y='LaunchSite', hue='Class', data=df)
plt.xlabel("FlightNumber")
plt.ylabel("LaunchSite")
plt.show
```

```python
sns.scatterplot(x='PayloadMass', y='LaunchSite', hue='Class', data=df)
plt.xlabel("PayloadMass")
plt.ylabel("LaunchSite")
plt.show
```

```python
df_orbit = df.groupby('Orbit')['Class'].mean()
df_orbit
```

```python
sns.barplot(x=df_orbit.index, y=df_orbit.values, hue = df_orbit.index)
plt.xlabel("Orbit")
plt.ylabel("Class")
plt.show()
```

```python
sns.scatterplot(x='FlightNumber', y='Orbit', hue='Class', data=df)
plt.xlabel("FlightNumber")
plt.ylabel("Orbit")
plt.show
```

```python
sns.scatterplot(x='PayloadMass', y='Orbit', hue='Class', data=df)
plt.xlabel("Payload")
plt.ylabel("Orbit")
plt.show
```

```python
year=[]
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
Extract_year()
df['Date'] = year
df.head()
```

```python
df_avgsuccess = df.groupby('Date')['Class'].mean()
df_avgsuccess
```

```python
sns.lineplot(x=df_avgsuccess.index, y=df_avgsuccess.values)
plt.xlabel("Date")
plt.ylabel("Average Success")
plt.show
```

```python
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]
features.head()
```

```python
features_one_hot = pd.get_dummies(features, columns=['Orbit', 'LaunchSite', 'LandingPad', 'Serial'])
features_one_hot.head()
```

```python
features_one_hot = features_one_hot.astype(float)
```

```python
features_one_hot.to_csv('dataset_part_3.csv', index=False)
```
